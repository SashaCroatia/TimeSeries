```python
import matplotlib.pyplot as plt
import numpy as np

# Figure dpi
dpi = 72
```

![](EMD-test-sliceoforbdiag.png)

In the image above, the verticle line is at $r=1.5$ and the horizontal line is the attractor $x$ of the logistic map at $r=1.5$. Say we sample three attractors from the logistic orbit diagram and three attractors from corresponding GP-fit orbit diagram for $r=1.5$. What is the EMD?

### Data from Experiments

```python
rsweep = [1.55,1.55,1.55]
x_f = [0.35483871, 0.35483871, 0.35483871]
x_GP = [0.35487263, 0.35487263, 0.35487263]
```

```python
fig = plt.figure(figsize=(1600/dpi,500/dpi),dpi=dpi)
ax = fig.add_subplot(121)

# Plot attractors vs r
ax.plot(rsweep, x_f, "kx", mew=2)
ax.plot(rsweep, x_GP, "bx", mew=2)
# Organize plot
ax.set_xlabel("r"), ax.set_ylabel("$x_n$")
ax.legend(labels=["Logistic Orbit", "GP-fit Orbit"]);

plt.subplots_adjust(wspace=0)
plt.show()
```

```python
#Compute the EMD:
from scipy.stats import wasserstein_distance
W_dist = wasserstein_distance(x_GP,x_f)
print(W_dist)
```
